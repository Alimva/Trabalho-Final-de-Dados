# <center> Os 20 principais países que geram mais energia com fontes renováveis ​​(2017)</center>

```python
import pandas as pd
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt


paises = pd.read_csv('D:/Users/lucie/Documents/Aula Data Analytic/main/database/top20CountriesPowerGeneration.csv')
paises
    
```

```python
df = pd.DataFrame(paises)
```

```python
df_total_p=df.iloc[:20,[0,5]]
df_total_p.sort_values(
    ["Total (TWh)"], axis=0,
    ascending=True, inplace=True)
df_total_p

X = list(df_total_p.iloc[:, 0])
Y = list(df_total_p.iloc[:, 1])
  
plt.figure(facecolor='gold')
plt.barh(X, Y, color='y')
plt.title("OS 20 PRINCIPAIS PAÍSES GERADORES DE ENERGIA RENOVÁVEL - 2017", fontweight='bold', color='w')
plt.xlabel("ENERGIA RENOVÁVEL", fontweight='bold',color='w' )
plt.ylabel("PAÍSES", fontweight='bold', color='w')
 
plt.show()
```

**Os 5 PAISES QUE MAIS PRODUZEM ENERGIA HIDRAÚLICA**

```python
df_Hidro_p=df.iloc[:5,[0,1]]
df_Hidro_p.sort_values(
    ["Hydro(TWh)"], axis=0,
    ascending=False, inplace=True)
df_Hidro_p
```

```python
df_Hidro_p=df.iloc[:5,[0,1]]
df_Hidro_p.sort_values(
    ["Hydro(TWh)"], axis=0,
    ascending=True, inplace=True)
df_Hidro_p

X = list(df_Hidro_p.iloc[:, 0])
Y = list(df_Hidro_p.iloc[:, 1])
  
plt.figure(facecolor='gray')
plt.barh(X, Y, color='black')
plt.title("OS 5 PAISES QUE MAIS PRODUZEM ENERGIA HIDRAÚLICA - 2017", fontweight='bold')
plt.xlabel("ENERGIA", fontweight='bold')
plt.ylabel("PAÍSES", fontweight='bold')
 
plt.show()
```

**Os 5 PAISES QUE MAIS PRODUZEM ENERGIA POR BIOCOMBUSTÍVEL**

```python
df_bio=df.iloc[:20,[0,2]]
df_bio.sort_values(
    ["Biofuel(TWh)"], axis=0,
    ascending=False, inplace=True)

df_bio_p=df_bio.iloc[:5,[0,1]]
df_bio_p
```

```python
X = list(df_bio_p.iloc[:, 0])
Y = list(df_bio_p.iloc[:, 1])

plt.figure(facecolor='b')
plt.bar(X, Y, color="g")
plt.title("Os 5 PAÍSES QUE MAIS PRODUZEM ENERGIA POR BIOCOMBUSTÍVEL", fontweight='bold', color='w')
plt.xlabel("ENERGIA POR BIOCOMBUSTÍVEL", fontweight='bold',color='w')
plt.ylabel("PAÍSES", fontweight='bold', color='w')
 
plt.show()
```
